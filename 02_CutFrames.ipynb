```python
def cut_images(lista_img, origem, destino):
    
    for i in lista_img:
        
        name = origem + str(i)
        
        image = Image.open(name) #abrindo imagem
        image_ar = np.array(image) #convertendo imagem em array
        new_img = image_ar[190:351,200:601,:] #cortando a imagem nas posições x[190:350] y[200,601]
        new_img = Image.fromarray(new_img) #Convertendo array em Image
        
        nome_destino = destino + str(i) #por algum motivo não salva
        new_img.save(nome_destino)
```

```python
def get_images_names(min_, max_):
    
    lista_img = []
    
    nome = 'frame_'
    ext = '.jpg'
    
    for i in range(min_, max_+1):
        
        if i < 10:
            num = '000'+str(i)
        elif i < 100:
            num = '00'+str(i)
        elif i < 1000:
            num = '0'+str(i)
        else:
            num = str(i)
        
        full_name = nome+num+ext
        lista_img.append(full_name)
        print(full_name)
    
    return lista_img
```

```python
origem = "Frames/Car/"
destino = "Frames/Car_cut/"
min_ = 0
max_ = 1500

lista_img = get_images_names(min_, max_)

cut_images(lista_img, origem, destino)
```
