```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import glob
```

```python
im1 = Image.open('Frames/Car_cut/frame_0180.jpg')
im2 = Image.open('Frames/Car_sub/frame_0180.jpg')
```

```python
im1 = np.array(im1)
im2 = np.array(im2)
```

```python
im1.shape
```

```python
def group(im1, im2, im3, im4):
    
    l, c, p = im1.shape

    result = np.zeros(shape=(l*2, c*2, p), dtype=np.uint8)

    for i in range(l):
        for j in range(c):

            result[i,j] = im1[i,j]

        for j in range(c):
            result[i,c+j] = im2[i,j]
            
    for i in range(l):
        for j in range(c):

            result[l+i,j] = im3[i,j]

        for j in range(c):
            result[l+i,c+j] = im4[i,j]

    return result
```

```python
def get_images(origem):
    
    img_array = []
    names = []
    for name in glob.glob(origem+'*.jpg'):
        
        img = Image.open(name)
        img = np.array(img)
        img_array.append(img)
        names.append(name)
        
    return img_array, names
```

```python
images1, names1 = get_images("Frames/Car_cut/")
images2, names2 = get_images("Frames/Car_sub/")
images3, names3 = get_images("Frames/Car_gray/")
images4, names4 = get_images("Frames/Car_sub_color/")
```

```python
final_images = []

for im1, im2, im3, im4 in zip(images1, images2, images3, images4):
    
    final_images.append(group(im1, im2, im3, im4))
      
```

```python
def save_images(images, nomes, destino):
    
    for nome, imagem in zip(nomes, images):
        
        destino_f = destino+nome.split('\\')[1]
        
        new_img = Image.fromarray(imagem) #Convertendo array em Image
        
        new_img.save(destino_f)
        print(f'Salvo: {destino_f}')
```

```python
save_images(final_images, names1, 'Frames/Car_full/')
```
