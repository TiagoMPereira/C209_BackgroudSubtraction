## Separação de frames
Este notebook é destinado à conversão de vídeos em frames

```python
# importando biblioteca opencv para tratamento de imagens
import cv2
```

```python
#Convertendo vídeos em frames e salvando na pasta de destino
#https://techtutorialsx.com/2021/04/29/python-opencv-splitting-video-frames/

def convert_frames(origem, destino):

    capture = cv2.VideoCapture(origem)

    frameNr = 0

    while (True):

        success, frame = capture.read()
        
        if frameNr < 10:
            frame_name = 'frame_000' + str(frameNr) + '.jpg'
        elif frameNr < 100:
            frame_name = 'frame_00' + str(frameNr) + '.jpg'
        elif frameNr < 1000:
            frame_name = 'frame_0' + str(frameNr) + '.jpg'
        else:
            frame_name = 'frame_' + str(frameNr) + '.jpg'

        destino_ = destino + frame_name

        if success:
            cv2.imwrite(destino_, frame)

        else:
            break

        print(f'Frame{frame_name} ok')
        frameNr = frameNr+1

    capture.release()
```

```python
origem = "Videos/Cars.mp4"
destino = "Frames/Car/"

convert_frames(origem, destino)
```
