```python
def color_sub(sub, normal):
    
    (l, c, p) = normal.shape
    
    img_ = np.zeros(shape = (l,c,p), dtype = np.uint8)
    
    for p_ in range(p): #para todas as páginas
        
        for l_ in range(l):
            
            for c_ in range (c):
                
                img_[l_, c_, p_] = int(normal[l_, c_, p_] * round(sub[l_, c_]/256))
        
    return img_
```

```python
def get_images(origem):
    
    img_array = []
    names = []
    for name in glob.glob(origem+'*.jpg'):
        
        img = Image.open(name)
        img = np.array(img)
        img_array.append(img)
        
        print(name)
        
        names.append(name.split('\\')[1])
        
    return img_array, names
```

```python
images_normal, normal_names = get_images("Frames/Car_cut/")
images_sub, sub_names = get_images("Frames/Car_sub/")
```

```python
def color_sub(sub_, normal_):
    
    images = []
    it = 0
    
    for sub, normal in zip(sub_, normal_):
    
        (l, c, p) = normal.shape

        img_ = np.zeros(shape = (l,c,p), dtype = np.uint8)

        for p_ in range(p): #para todas as páginas

            for l_ in range(l):

                for c_ in range (c):

                    img_[l_, c_, p_] = int(normal[l_, c_, p_] * round(sub[l_, c_]/256))
        
        images.append(img_)
        print(it)
        it+=1
        
    return images
```

```python
images_color = color_sub(images_sub, images_normal)
```

```python
def save_sub_images(images, nomes, destino):
    
    
    for nome, imagem in zip(nomes, images):
        
        destino_f = destino+nome
        
        new_img = Image.fromarray(imagem) #Convertendo array em Image
        
        new_img.save(destino_f)
        print(f'Salvo: {destino_f}')
```

```python
destino = "Frames/Car_sub_color/"
save_sub_images(images_color, normal_names, destino)
```
